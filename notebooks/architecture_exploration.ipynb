### Sobel Edge Detection

```python
img = cv2.imread('architecture_cache/architecture_ex_01.jpg')
plt.imshow(img)
plt.show()
```

Convert the image to greyscale

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(img_gray, cmap='gray')
plt.show()
```

Blur the image using Gaussian Blur. 
This will remove smaller redundant detail not neccessarily critical to the major edge detection.

```python
img_blur = cv2.GaussianBlur(img_gray, (3,3), 0)
plt.imshow(img_blur, cmap='gray')
plt.show()
```

Perform Sobel edge detection.

```python
sobel_x = cv2.Sobel(src=img_blur, ddepth=cv2.CV_64F, dx=1, dy=0, ksize=5) # Sobel Edge Detection on the X axis
sobel_y = cv2.Sobel(src=img_blur, ddepth=cv2.CV_64F, dx=0, dy=1, ksize=5) # Sobel Edge Detection on the Y axis
sobel_xy = cv2.Sobel(src=img_blur, ddepth=cv2.CV_64F, dx=1, dy=1, ksize=5) # Combined X and Y Sobel Edge Detection
```

```python
plt.imshow(sobel_x, cmap='gray')
plt.show()
```

```python
plt.imshow(sobel_y, cmap='gray')
plt.show()
```

```python
plt.imshow(sobel_xy, cmap='gray')
plt.show()
```

### Canny Edge Detection Alternative

```python
edges = cv2.Canny(image=img_blur, threshold1=100, threshold2=200)
plt.imshow(edges, cmap='gray')
plt.show()
```
